## Step 4 — Launch App

> 🔑 Make sure you've added your ngrok token to **Colab Secrets** (left sidebar → key icon) with the name `NGROK_TOKEN` before running this cell.

```python
from google.colab import userdata
from pyngrok import ngrok
import nest_asyncio
import uvicorn
from main import app

# Auth
ngrok.set_auth_token(userdata.get('NGROK_TOKEN'))

# Allow asyncio inside Jupyter
nest_asyncio.apply()

# Create tunnel
public_url = ngrok.connect(8000)
print('=' * 60)
print(f'  ✅ App is running!')
print(f'  👉 Open this URL: {public_url}')
print('=' * 60)
print('  Share this URL with your team.')
print('  The URL will change if you restart the notebook.')
print('=' * 60)

# Run server
config = uvicorn.Config(app, host='0.0.0.0', port=8000, log_level='warning')
server = uvicorn.Server(config)
await server.serve()
```

## Stop — Kill Tunnel

Run this cell when you're done to close the public URL.

```python
ngrok.kill()
print('✅ Tunnel closed.')
```
